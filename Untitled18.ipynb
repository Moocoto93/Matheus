```python
# Instalação de dependências (descomente se necessário)
!pip install dash dash-bootstrap-components pandas plotly pyngrok statsmodels scikit-learn

import os
import time
import logging
from threading import Thread

import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import poisson
from datetime import datetime
import plotly.express as px
import plotly.graph_objects as go
import dash
from dash import dcc, html, dash_table, no_update
from dash.dependencies import Input, Output, State
import requests
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from concurrent.futures import ThreadPoolExecutor
from sklearn.ensemble import GradientBoostingRegressor
from pyngrok import ngrok

# Configuração básica do logging
logging.basicConfig(level=logging.INFO)

# Configuração do Telegram (substitua pelos seus valores reais)
TELEGRAM_BOT_TOKEN = "SEU_BOT_TOKEN_AQUI"  # Exemplo: "123456789:ABCDEF..."
TELEGRAM_CHAT_ID = "SEU_CHAT_ID_AQUI"        # Exemplo: "-123456789"

# =============================================================================
# CONFIGURAÇÕES E DADOS
# =============================================================================
LEAGUE_URLS = {
    "Alemanha": "https://www.football-data.co.uk/mmz4281/2425/D2.csv",
    "Espanha": "https://www.football-data.co.uk/mmz4281/2425/SP1.csv",
    "França": "https://www.football-data.co.uk/mmz4281/2425/F1.csv",
    "Holanda": "https://www.football-data.co.uk/mmz4281/2425/N1.csv",
    "Inglaterra": "https://www.football-data.co.uk/mmz4281/2425/E0.csv",
    "Itália": "https://www.football-data.co.uk/mmz4281/2425/I1.csv",
    "Portugal": "https://www.football-data.co.uk/mmz4281/2425/P1.csv",
    "Turquia": "https://www.football-data.co.uk/mmz4281/2425/T1.csv"
}

# =============================================================================
# FUNÇÕES DE CARREGAMENTO E TRATAMENTO DE DADOS
# =============================================================================
def load_data(league: str) -> pd.DataFrame:
    url = LEAGUE_URLS[league]
    try:
        df = pd.read_csv(url)
        df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')
    except Exception as e:
        logging.error(f"Erro ao carregar os dados: {e}")
        raise RuntimeError(f"Erro ao carregar os dados: {e}")
    required_cols = ['FTHG', 'FTAG', 'HomeTeam', 'AwayTeam']
    for col in required_cols:
        if col not in df.columns:
            raise KeyError(f"Coluna '{col}' não encontrada no dataset.")
    for col in ['HC', 'AC', 'HS', 'AS', 'HST', 'AST']:
        if col not in df.columns:
            logging.warning(f"Coluna '{col}' não encontrada no dataset.")
            df[col] = 0  # Preencher com 0 para evitar erros
    return df.dropna(subset=required_cols).copy()

def get_all_teams(df: pd.DataFrame) -> list:
    return pd.concat([df['HomeTeam'], df['AwayTeam']]).unique().tolist()

def get_classification_table(df: pd.DataFrame) -> pd.DataFrame:
    teams = df['HomeTeam'].unique()
    classification = []
    corners_available = all(col in df.columns for col in ['HC', 'AC'])
    shots_available = all(col in df.columns for col in ['HS', 'AS', 'HST', 'AST'])

    for team in teams:
        home = df[df['HomeTeam'] == team]
        away = df[df['AwayTeam'] == team]
        played = len(home) + len(away)
        wins = (home['FTHG'] > home['FTAG']).sum() + (away['FTAG'] > away['FTHG']).sum()
        draws = (home['FTHG'] == home['FTAG']).sum() + (away['FTAG'] == away['FTHG']).sum()
        losses = (home['FTHG'] < home['FTAG']).sum() + (away['FTAG'] < away['FTHG']).sum()
        goals_for = home['FTHG'].sum() + away['FTAG'].sum()
        goals_against = home['FTAG'].sum() + away['FTHG'].sum()
        goal_diff = goals_for - goals_against
        points = wins * 3 + draws
        offensive_strength = goals_for / played if played > 0 else 0
        defensive_strength = goals_against / played if played > 0 else 0

        entry = {
            'Team': team,
            'P': played,
            'W': wins,
            'D': draws,
            'L': losses,
            'GF': goals_for,
            'GA': goals_against,
            'GD': goal_diff,
            'Pts': points,
            'Offensive Strength': offensive_strength,
            'Defensive Strength': defensive_strength
        }

        if shots_available:
            total_shots = home['HS'].sum() + away['AS'].sum()
            total_shots_on_target = home['HST'].sum() + away['AST'].sum()
            avg_shots_on_target = total_shots_on_target / played if played > 0 else 0
            entry['Total Shots'] = total_shots
            entry['Avg Shots on Target'] = avg_shots_on_target

        if corners_available:
            corners_for = home['HC'].sum() + away['AC'].sum()
            corners_against = home['AC'].sum() + away['HC'].sum()
            entry['Avg Corners For'] = corners_for / played if played > 0 else 0
            entry['Avg Corners Against'] = corners_against / played if played > 0 else 0

        classification.append(entry)

    table = pd.DataFrame(classification).sort_values(by=['Pts', 'GD', 'GF'], ascending=False).reset_index(drop=True)
    table.index += 1
    return table

def add_classification_features(df: pd.DataFrame) -> pd.DataFrame:
    classif = get_classification_table(df)
    classif["off_strength"] = classif["GF"] / classif["P"]
    classif["def_strength"] = classif["GA"] / classif["P"]
    off_map = dict(zip(classif["Team"], classif["off_strength"]))
    def_map = dict(zip(classif["Team"], classif["def_strength"]))
    df["home_off_strength"] = df["HomeTeam"].map(off_map)
    df["home_def_strength"] = df["HomeTeam"].map(def_map)
    df["away_off_strength"] = df["AwayTeam"].map(off_map)
    df["away_def_strength"] = df["AwayTeam"].map(def_map)
    if "Avg Corners For" in classif.columns:
        corners_for_map = dict(zip(classif["Team"], classif["Avg Corners For"]))
        corners_against_map = dict(zip(classif["Team"], classif["Avg Corners Against"]))
        df["home_avg_corners_for"] = df["HomeTeam"].map(corners_for_map)
        df["home_avg_corners_against"] = df["HomeTeam"].map(corners_against_map)
        df["away_avg_corners_for"] = df["AwayTeam"].map(corners_for_map)
        df["away_avg_corners_against"] = df["AwayTeam"].map(corners_against_map)
    return df

# =============================================================================
# FUNÇÕES DE PREDIÇÃO
# =============================================================================
def predict_asian_handicap(lambda_home: float, lambda_away: float, handicap: float, max_goals: int = 10) -> float:
    prob = 0
    for i in range(max_goals + 1):
        for j in range(max_goals + 1):
            diff_adjusted = (i - j) + handicap
            if diff_adjusted > 0:
                prob += poisson.pmf(i, lambda_home) * poisson.pmf(j, lambda_away)
    return prob

def predict_match_outcomes(lambda_home: float, lambda_away: float, max_goals: int = 10) -> tuple:
    home_win = draw = away_win = btts = over_2_5 = 0
    for i in range(max_goals + 1):
        for j in range(max_goals + 1):
            p = poisson.pmf(i, lambda_home) * poisson.pmf(j, lambda_away)
            if i > j:
                home_win += p
            elif i == j:
                draw += p
            else:
                away_win += p
            if i >= 1 and j >= 1:
                btts += p
            if (i + j) > 2.5:
                over_2_5 += p
    return home_win, draw, away_win, btts, over_2_5

# =============================================================================
# FUNÇÕES DE CÁLCULO DE FORMA E ESTATÍSTICAS
# =============================================================================
def compute_rolling_features(df: pd.DataFrame, team: str, window: int = 5) -> dict:
    home_df = df[df['HomeTeam'] == team].sort_values('Date')
    if not home_df.empty:
        home_rolling = home_df[['FTHG', 'FTAG']].rolling(window=window, min_periods=1).mean().iloc[-1].to_dict()
    else:
        home_rolling = {'FTHG': np.nan, 'FTAG': np.nan}
    away_df = df[df['AwayTeam'] == team].sort_values('Date')
    if not away_df.empty:
        away_rolling = away_df[['FTAG', 'FTHG']].rolling(window=window, min_periods=1).mean().iloc[-1].to_dict()
    else:
        away_rolling = {'FTAG': np.nan, 'FTHG': np.nan}
    return {
        'home': {'rolling_GF': home_rolling.get('FTHG', np.nan), 'rolling_GA': home_rolling.get('FTAG', np.nan)},
        'away': {'rolling_GF': away_rolling.get('FTAG', np.nan), 'rolling_GA': away_rolling.get('FTHG', np.nan)}
    }

def compute_team_form(df: pd.DataFrame, window: int = 5) -> tuple:
    teams = pd.concat([df['HomeTeam'], df['AwayTeam']]).unique()
    home_stats, away_stats = {}, {}
    for team in teams:
        team_home = df[df['HomeTeam'] == team].sort_values('Date').tail(window)
        team_away = df[df['AwayTeam'] == team].sort_values('Date').tail(window)
        gf_home = team_home['FTHG'].sum() / min(len(team_home), window) if not team_home.empty else 0
        ga_home = team_home['FTAG'].sum() / min(len(team_home), window) if not team_home.empty else 0
        gf_away = team_away['FTAG'].sum() / min(len(team_away), window) if not team_away.empty else 0
        ga_away = team_away['FTHG'].sum() / min(len(team_away), window) if not team_away.empty else 0
        home_stats[team] = {'gf': gf_home, 'ga': ga_home}
        away_stats[team] = {'gf': gf_away, 'ga': ga_away}
    return home_stats, away_stats

def add_form_features(df: pd.DataFrame, home_stats: dict, away_stats: dict, classification_table: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df['home_recent_GF'] = df['HomeTeam'].map(lambda x: home_stats.get(x, {}).get('gf', 0))
    df['home_recent_GA'] = df['HomeTeam'].map(lambda x: home_stats.get(x, {}).get('ga', 0))
    df['away_recent_GF'] = df['AwayTeam'].map(lambda x: away_stats.get(x, {}).get('gf', 0))
    df['away_recent_GA'] = df['AwayTeam'].map(lambda x: away_stats.get(x, {}).get('ga', 0))
    df['home_advantage'] = 1
    df['opponent_defense'] = df['AwayTeam'].map(lambda x: away_stats.get(x, {}).get('ga', 0))
    df['opponent_offense'] = df['HomeTeam'].map(lambda x: home_stats.get(x, {}).get('gf', 0))
    df = df.merge(classification_table[['Team', 'Offensive Strength', 'Defensive Strength']], left_on='HomeTeam', right_on='Team', how='left')
    df = df.rename(columns={'Offensive Strength': 'home_off_strength', 'Defensive Strength': 'home_def_strength'})
    df = df.merge(classification_table[['Team', 'Offensive Strength', 'Defensive Strength']], left_on='AwayTeam', right_on='Team', how='left')
    df = df.rename(columns={'Offensive Strength': 'away_off_strength', 'Defensive Strength': 'away_def_strength'})
    df = df.drop(columns=['Team_x', 'Team_y'])
    if 'Avg Corners For' in classification_table.columns:
        df = df.merge(classification_table[['Team', 'Avg Corners For', 'Avg Corners Against']], left_on='HomeTeam', right_on='Team', how='left')
        df = df.rename(columns={'Avg Corners For': 'home_avg_corners_for', 'Avg Corners Against': 'home_avg_corners_against'})
        df = df.merge(classification_table[['Team', 'Avg Corners For', 'Avg Corners Against']], left_on='AwayTeam', right_on='Team', how='left')
        df = df.rename(columns={'Avg Corners For': 'away_avg_corners_for', 'Avg Corners Against': 'away_avg_corners_against'})
        df = df.drop(columns=['Team_x', 'Team_y'])
    return df

def optimize_window_size(df: pd.DataFrame, windows: list = [3, 5, 10]) -> int:
    return 5

def evaluate_poisson_model(df_form: pd.DataFrame, formula: str, target: str, n_splits: int = 5) -> float:
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    mse_scores = []
    for train_idx, test_idx in kf.split(df_form):
        train, test = df_form.iloc[train_idx], df_form.iloc[test_idx]
        model = smf.glm(formula=formula, data=train, family=sm.families.Poisson()).fit()
        preds = model.predict(test)
        mse_scores.append(mean_squared_error(test[target], preds))
    mse_mean = np.mean(mse_scores)
    logging.info(f"MSE para {target}: {mse_mean:.2f}")
    return mse_mean

def fit_poisson_models_with_form(df_form: pd.DataFrame) -> tuple:
    formula_home = ("FTHG ~ home_recent_GF + home_recent_GA + home_advantage + opponent_defense + "
                    "home_off_strength + away_def_strength")
    formula_away = ("FTAG ~ away_recent_GF + away_recent_GA + opponent_offense + "
                    "away_off_strength + home_def_strength")
    if 'home_avg_corners_for' in df_form.columns:
        formula_home += " + home_avg_corners_for + away_avg_corners_against"
        formula_away += " + away_avg_corners_for + home_avg_corners_against"

    def train_model(formula, data, target):
        return smf.glm(formula=formula, data=data, family=sm.families.Poisson()).fit()

    with ThreadPoolExecutor(max_workers=2) as executor:
        future_home = executor.submit(train_model, formula_home, df_form, 'FTHG')
        future_away = executor.submit(train_model, formula_away, df_form, 'FTAG')
        model_home = future_home.result()
        model_away = future_away.result()

    mse_home = evaluate_poisson_model(df_form, formula_home, 'FTHG')
    mse_away = evaluate_poisson_model(df_form, formula_away, 'FTAG')
    logging.info(f"MSE Home: {mse_home:.2f} | MSE Away: {mse_away:.2f}")

    home_stats, away_stats = compute_team_form(df_form)
    return model_home, model_away, home_stats, away_stats

def fit_poisson_models_corners(df_form: pd.DataFrame) -> tuple:
    if 'HC' not in df_form.columns or 'AC' not in df_form.columns:
        raise KeyError("Colunas de escanteios (HC/AC) não encontradas.")
    formula_home = "HC ~ home_recent_GF + home_recent_GA + home_advantage + opponent_defense + home_off_strength + away_def_strength"
    if 'home_avg_corners_for' in df_form.columns:
        formula_home += " + home_avg_corners_for + away_avg_corners_against"
    formula_away = "AC ~ away_recent_GF + away_recent_GA + opponent_offense + away_off_strength + home_def_strength"
    if 'away_avg_corners_for' in df_form.columns:
        formula_away += " + away_avg_corners_for + home_avg_corners_against"
    model_corners_home = smf.glm(formula=formula_home, data=df_form, family=sm.families.Poisson()).fit()
    model_corners_away = smf.glm(formula=formula_away, data=df_form, family=sm.families.Poisson()).fit()
    return model_corners_home, model_corners_away

def calculate_corners_probabilities(lambda_total_corners: float, threshold: float = 8.5) -> tuple:
    prob_over = 1 - poisson.cdf(threshold, lambda_total_corners)
    prob_under = poisson.cdf(threshold, lambda_total_corners)
    return prob_over, prob_under

def odds(prob: float) -> float:
    return 1 / prob if prob > 0 else float('inf')

def send_telegram_notification(message: str, chat_id: str, bot_token: str):
    url = f"https://api.telegram.org/bot{bot_token}/sendMessage"
    payload = {"chat_id": chat_id, "text": message}
    try:
        response = requests.post(url, json=payload)
        response.raise_for_status()
    except requests.RequestException as e:
        logging.error(f"Erro ao enviar mensagem para o Telegram: {e}")

def similar_opponent_performance(df: pd.DataFrame, team: str, classification_table: pd.DataFrame, reference_team: str) -> str:
    if reference_team not in classification_table['Team'].values:
        return f"Dados insuficientes para análise com {reference_team}."
    classification_sorted = classification_table.sort_values(by=['Pts', 'GD', 'GF'], ascending=False).reset_index(drop=True)
    ref_rank = classification_sorted[classification_sorted['Team'] == reference_team].index[0] + 1
    similar_ranks = classification_sorted.iloc[max(0, ref_rank-3): min(len(classification_sorted), ref_rank+2)]
    similar_teams = similar_ranks['Team'].tolist()
    team_matches = df[(df['HomeTeam'] == team) | (df['AwayTeam'] == team)].copy()
    team_matches['Opponent'] = team_matches.apply(lambda row: row['AwayTeam'] if row['HomeTeam'] == team else row['HomeTeam'], axis=1)
    similar_matches = team_matches[team_matches['Opponent'].isin(similar_teams)]
    if similar_matches.empty:
        return "Nenhum jogo contra adversários similares encontrado."
    similar_matches['GoalDiff'] = similar_matches.apply(lambda row: (row['FTHG'] - row['FTAG']) if row['HomeTeam'] == team else (row['FTAG'] - row['FTHG']), axis=1)
    avg_diff = similar_matches['GoalDiff'].mean()
    return f"Média Diferença de Gols: {avg_diff:.2f}"

def compute_similarity_table(df: pd.DataFrame, team: str, classification_table: pd.DataFrame, reference_team: str, n_matches: int = 7) -> pd.DataFrame:
    if reference_team not in classification_table['Team'].values:
        return pd.DataFrame({"Mensagem": [f"Dados insuficientes para análise com {reference_team}."]})
    ref_points = classification_table[classification_table['Team'] == reference_team]['Pts'].iloc[0]
    max_diff = (classification_table['Pts'] - ref_points).abs().max() or 1
    team_matches = df[(df['HomeTeam'] == team) | (df['AwayTeam'] == team)].copy()
    team_matches = team_matches.sort_values('Date', ascending=False)
    recent_matches = team_matches.head(n_matches)
    def get_opponent(row):
        return row['AwayTeam'] if row['HomeTeam'] == team else row['HomeTeam']
    recent_matches['Opponent'] = recent_matches.apply(get_opponent, axis=1)
    def get_venue(row):
        return "Casa" if row['HomeTeam'] == team else "Fora"
    recent_matches['Venue'] = recent_matches.apply(get_venue, axis=1)
    def compute_sim_score(opponent):
        opp_points = classification_table[classification_table['Team'] == opponent]['Pts'].iloc[0]
        return abs(opp_points - ref_points) / max_diff
    recent_matches['SimScore'] = recent_matches['Opponent'].apply(compute_sim_score)
    def get_result(row):
        if row['HomeTeam'] == team:
            if row['FTHG'] > row['FTAG']:
                return "Vitória"
            elif row['FTHG'] == row['FTAG']:
                return "Empate"
            else:
                return "Derrota"
        else:
            if row['FTAG'] > row['FTHG']:
                return "Vitória"
            elif row['FTAG'] == row['FTHG']:
                return "Empate"
            else:
                return "Derrota"
    recent_matches['Result'] = recent_matches.apply(get_result, axis=1)
    def get_score(row):
        if row['HomeTeam'] == team:
            return f"{row['FTHG']} - {row['FTAG']}"
        else:
            return f"{row['FTAG']} - {row['FTHG']}"
    recent_matches['Score'] = recent_matches.apply(get_score, axis=1)
    recent_matches['Date'] = recent_matches['Date'].dt.strftime("%d/%m/%Y")
    recent_matches = recent_matches.sort_values('SimScore', ascending=True)
    return recent_matches[['Date', 'Opponent', 'Venue', 'Result', 'Score', 'SimScore']].reset_index(drop=True)

# =============================================================================
# FUNÇÕES DE ENSEMBLE
# =============================================================================
def train_gradient_boosting_model(df_form: pd.DataFrame) -> tuple:
    df_copy = df_form.copy()
    df_copy['total_goals'] = df_copy['FTHG'] + df_copy['FTAG']
    features = ['home_recent_GF', 'home_recent_GA', 'away_recent_GF', 'away_recent_GA',
                'home_advantage', 'opponent_defense', 'opponent_offense',
                'home_off_strength', 'home_def_strength', 'away_off_strength', 'away_def_strength']
    if 'home_avg_corners_for' in df_copy.columns:
        features.extend(['home_avg_corners_for', 'home_avg_corners_against', 'away_avg_corners_for', 'away_avg_corners_against'])
    X = df_copy[features]
    y = df_copy['total_goals']
    gb_model = GradientBoostingRegressor(random_state=42)
    gb_model.fit(X, y)
    return gb_model, features

def ensemble_prediction(poisson_pred: float, gb_pred: float, weight: float = 0.5) -> float:
    return weight * poisson_pred + (1 - weight) * gb_pred

# =============================================================================
# FUNÇÃO: Performance por Nível de Adversário
# =============================================================================
def performance_by_opponent_level(df: pd.DataFrame, team: str, classification_table: pd.DataFrame, margin: float = 3) -> pd.DataFrame:
    team_points = classification_table[classification_table['Team'] == team]['Pts'].iloc[0]
    team_matches = df[(df['HomeTeam'] == team) | (df['AwayTeam'] == team)].copy()
    def get_category(row):
        opponent = row['AwayTeam'] if row['HomeTeam'] == team else row['HomeTeam']
        opp_points = classification_table[classification_table['Team'] == opponent]['Pts'].iloc[0]
        if opp_points >= team_points + margin:
            return "Superior"
        elif opp_points <= team_points - margin:
            return "Inferior"
        else:
            return "Similar"
    team_matches['Category'] = team_matches.apply(get_category, axis=1)
    def calc_metrics(row):
        if row['HomeTeam'] == team:
            goals_scored = row['FTHG']
            goals_conceded = row['FTAG']
            shot_eff = (row['HST']/row['HS']) if ('HS' in row and row['HS'] > 0) else np.nan
        else:
            goals_scored = row['FTAG']
            goals_conceded = row['FTHG']
            shot_eff = (row['AST']/row['AS']) if ('AS' in row and row['AS'] > 0) else np.nan
        return pd.Series({
            'Goals Scored': goals_scored,
            'Goals Conceded': goals_conceded,
            'Goal Diff': goals_scored - goals_conceded,
            'Shot Efficiency': shot_eff
        })
    metrics = team_matches.apply(calc_metrics, axis=1)
    team_matches = pd.concat([team_matches, metrics], axis=1)
    group = team_matches.groupby('Category').agg(
        Games=('Category', 'count'),
        Avg_Goals_Scored=('Goals Scored', 'mean'),
        Avg_Goals_Conceded=('Goals Conceded', 'mean'),
        Avg_Goal_Diff=('Goal Diff', 'mean'),
        Avg_Shot_Efficiency=('Shot Efficiency', 'mean')
    ).reset_index()
    return group

# =============================================================================
# DASHBOARD
# =============================================================================
app = dash.Dash(__name__, suppress_callback_exceptions=True)
server = app.server  # Necessário para hospedagem online com gunicorn
app.title = "Dashboard de Análises de Futebol"

app.layout = html.Div([
    html.H1("Dashboard de Apostas - Análise de Dados de Futebol"),
    html.Div([
        html.Label("Selecione o Campeonato:"),
        dcc.Dropdown(
            id='league-dropdown',
            options=[{'label': key, 'value': key} for key in sorted(LEAGUE_URLS.keys())],
            value="Inglaterra",
            clearable=False,
            style={'width': '50%'}
        )
    ], style={'padding': '10px'}),
    html.Br(),
    html.Div([
        dash_table.DataTable(
            id='classification-table',
            columns=[{"name": col, "id": col} for col in ['Team', 'P', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts',
                                                          'Total Shots', 'Avg Shots on Target']],
            data=[],
            page_action='none',
            style_table={'overflowX': 'auto'},
            style_cell={'textAlign': 'center'},
            style_header={'backgroundColor': 'lightgrey', 'fontWeight': 'bold'}
        )
    ], style={'padding': '10px'}),
    html.Br(),
    html.Div([dcc.Graph(id='ranking-graph')], style={'padding': '10px'}),
    html.Hr(),
    html.H2("Análise de Jogo"),
    html.Div([
        html.Label("Time Casa:"),
        dcc.Dropdown(id='home-team-dropdown', options=[], placeholder="Selecione o Time Casa", style={'width': '50%'})
    ], style={'padding': '10px'}),
    html.Div([
        html.Label("Time Visitante:"),
        dcc.Dropdown(id='away-team-dropdown', options=[], placeholder="Selecione o Time Visitante", style={'width': '50%'})
    ], style={'padding': '10px'}),
    html.Div([
        html.Label("Handicap:"),
        dcc.Input(id='handicap-input', type='number', value=0, step=0.25, style={'width': '20%'})
    ], style={'padding': '10px'}),
    html.Div([
        html.Label("Threshold Over Escanteios:"),
        dcc.Slider(id='corners-threshold', min=5.5, max=12.5, step=0.5, value=8.5, marks={i: str(i) for i in range(6,13)})
    ], style={'padding': '10px'}),
    html.Div([
        html.Label("Probabilidade Mínima para Recomendação:"),
        dcc.Slider(id='prob-threshold', min=0.3, max=0.7, step=0.05, value=0.5, marks={i/10: str(i/10) for i in range(3,8)})
    ], style={'padding': '10px'}),
    html.Button("Prever Jogo", id='predict-match-button', n_clicks=0, style={'margin': '10px'}),
    dcc.Loading(
        id="loading-analysis",
        type="default",
        children=[
            html.Div(id='match-analysis-output', style={'padding': '10px'}),
            html.Div([dcc.Graph(id='heatmap-graph')], style={'padding': '10px'})
        ]
    ),
    html.Hr(),
    html.H2("Tendências de Gols por Jogo"),
    dcc.Loading(
        id="loading-trends",
        type="default",
        children=[
            html.Div([
                dcc.Graph(id='home-goals-trend-graph'),
                dcc.Graph(id='away-goals-trend-graph')
            ], style={'padding': '10px'}),
            html.Div([dcc.Graph(id='form-comparison-graph')], style={'padding': '10px'})
        ]
    ),
    html.Hr(),
    html.H2("Odds Preditadas e Recomendações"),
    dash_table.DataTable(
        id='odds-table',
        columns=[
            {"name": "Evento", "id": "Evento"},
            {"name": "Probabilidade", "id": "Probabilidade"},
            {"name": "Odds Calculadas", "id": "Odds Calculadas"},
            {"name": "Recomendação", "id": "Recomendação"}
        ],
        data=[],
        style_cell={'textAlign': 'center'},
        style_header={'backgroundColor': 'lightgrey', 'fontWeight': 'bold'},
        style_data_conditional=[
            {'if': {'filter_query': '{Recomendação} = "Aposta recomendada"', 'column_id': 'Recomendação'},
             'backgroundColor': '#d4edda', 'color': 'green'},
            {'if': {'filter_query': '{Recomendação} = "Atenção"', 'column_id': 'Recomendação'},
             'backgroundColor': '#fff3cd', 'color': 'orange'},
            {'if': {'filter_query': '{Recomendação} = "Cautela"', 'column_id': 'Recomendação'},
             'backgroundColor': '#f8d7da', 'color': 'red'}
        ]
    ),
    html.Button("Exportar Odds como CSV", id='export-odds-btn', style={'margin': '10px'}),
    dcc.Download(id="download-odds"),
    html.Br(),
    html.Hr(),
    html.H2("Tabela de Similaridade com Adversários Similares"),
    dash_table.DataTable(
        id='similarity-table',
        columns=[
            {"name": "Date", "id": "Date"},
            {"name": "Adversário", "id": "Opponent"},
            {"name": "Venue", "id": "Venue"},
            {"name": "Result", "id": "Result"},
            {"name": "Score", "id": "Score"},
            {"name": "SimScore", "id": "SimScore"}
        ],
        data=[],
        page_action='none',
        style_table={'overflowX': 'auto'},
        style_cell={'textAlign': 'center'},
        style_header={'backgroundColor': 'lightgrey', 'fontWeight': 'bold'},
        style_data_conditional=[
            {'if': {'filter_query': '{Result} = "Vitória"', 'column_id': 'Result'},
             'backgroundColor': '#d4edda', 'color': 'green'},
            {'if': {'filter_query': '{Result} = "Empate"', 'column_id': 'Result'},
             'backgroundColor': '#fff3cd', 'color': 'orange'},
            {'if': {'filter_query': '{Result} = "Derrota"', 'column_id': 'Result'},
             'backgroundColor': '#f8d7da', 'color': 'red'}
        ]
    ),
    html.Button("Exportar Similaridade como CSV", id='export-similarity-btn', style={'margin': '10px'}),
    dcc.Download(id="download-similarity"),
    html.Hr(),
    html.H2("Performance por Nível de Adversário"),
    dash_table.DataTable(
        id='performance-table',
        columns=[
            {"name": "Category", "id": "Category"},
            {"name": "Games", "id": "Games"},
            {"name": "Avg Goals Scored", "id": "Avg_Goals_Scored"},
            {"name": "Avg Goals Conceded", "id": "Avg_Goals_Conceded"},
            {"name": "Avg Goal Diff", "id": "Avg_Goal_Diff"},
            {"name": "Avg Shot Efficiency", "id": "Avg_Shot_Efficiency"}
        ],
        data=[],
        style_cell={'textAlign': 'center'},
        style_header={'backgroundColor': 'lightgrey', 'fontWeight': 'bold'},
        style_table={'overflowX': 'auto'}
    )
])

# =============================================================================
# CALLBACKS
# =============================================================================
@app.callback(
    [Output('home-team-dropdown', 'options'),
     Output('away-team-dropdown', 'options')],
    [Input('league-dropdown', 'value')]
)
def update_team_dropdowns(selected_league: str) -> tuple:
    df = load_data(selected_league)
    teams = sorted(get_all_teams(df))
    options = [{'label': team, 'value': team} for team in teams]
    return options, options

@app.callback(
    [Output('classification-table', 'data'),
     Output('ranking-graph', 'figure')],
    [Input('league-dropdown', 'value')]
)
def update_dashboard(selected_league: str) -> tuple:
    df = load_data(selected_league)
    classification_table = get_classification_table(df)
    fig = px.bar(classification_table, x='Team', y='Pts',
                 title=f"Tabela de Classificação - {selected_league}",
                 labels={'Team': 'Time', 'Pts': 'Pontos'}, text='Pts')
    fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
    fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
    return classification_table.to_dict('records'), fig

@app.callback(
    [Output('match-analysis-output', 'children'),
     Output('odds-table', 'data'),
     Output('similarity-table', 'data'),
     Output('heatmap-graph', 'figure'),
     Output('performance-table', 'data')],
    [Input('predict-match-button', 'n_clicks'),
     Input("export-odds-btn", "n_clicks"),
     Input("export-similarity-btn", "n_clicks")],
    [State('league-dropdown', 'value'),
     State('home-team-dropdown', 'value'),
     State('away-team-dropdown', 'value'),
     State('handicap-input', 'value'),
     State('corners-threshold', 'value'),
     State('prob-threshold', 'value'),
     State('odds-table', 'data'),
     State('similarity-table', 'data')]
)
def analyze_match(n_clicks, export_odds_clicks, export_similarity_clicks,
                  selected_league, home_team, away_team, handicap, corners_threshold, prob_threshold,
                  odds_data, similarity_data):
    try:
        if n_clicks == 0 or not home_team or not away_team:
            return "Selecione os times e clique em 'Prever Jogo' para ver os resultados.", no_update, no_update, go.Figure(), no_update

        # Validação do handicap
        if handicap is None or not isinstance(handicap, (int, float)):
            return "Erro: Handicap inválido. Insira um número válido.", no_update, no_update, go.Figure(), no_update

        df = load_data(selected_league)
        if len(df[df['HomeTeam'] == home_team]) < 3 or len(df[df['AwayTeam'] == away_team]) < 3:
            warning = "Aviso: Um ou ambos os times têm menos de 3 jogos. As previsões podem ser menos precisas.\n\n"
        else:
            warning = ""

        classification_table = get_classification_table(df)
        best_window = optimize_window_size(df)
        home_stats, away_stats = compute_team_form(df, window=best_window)
        df_form = add_form_features(df, home_stats, away_stats, classification_table)
        model_home, model_away, _, _ = fit_poisson_models_with_form(df_form)

        pred_df = pd.DataFrame({
            'HomeTeam': [home_team],
            'AwayTeam': [away_team],
            'home_recent_GF': [home_stats.get(home_team, {}).get('gf', 0)],
            'home_recent_GA': [home_stats.get(home_team, {}).get('ga', 0)],
            'away_recent_GF': [away_stats.get(away_team, {}).get('gf', 0)],
            'away_recent_GA': [away_stats.get(away_team, {}).get('ga', 0)],
            'home_advantage': [1],
            'opponent_defense': [away_stats.get(away_team, {}).get('ga', 0)],
            'opponent_offense': [home_stats.get(home_team, {}).get('gf', 0)]
        })
        home_row = classification_table[classification_table['Team'] == home_team]
        away_row = classification_table[classification_table['Team'] == away_team]
        if home_row.empty or away_row.empty:
            raise ValueError("Não foi possível obter informações da classificação para um dos times.")
        home_off_strength = home_row['Offensive Strength'].iloc[0]
        home_def_strength = home_row['Defensive Strength'].iloc[0]
        away_off_strength = away_row['Offensive Strength'].iloc[0]
        away_def_strength = away_row['Defensive Strength'].iloc[0]

        pred_df['home_off_strength'] = home_off_strength
        pred_df['home_def_strength'] = home_def_strength
        pred_df['away_off_strength'] = away_off_strength
        pred_df['away_def_strength'] = away_def_strength
        if 'Avg Corners For' in classification_table.columns:
            pred_df['home_avg_corners_for'] = home_row['Avg Corners For'].iloc[0]
            pred_df['home_avg_corners_against'] = home_row['Avg Corners Against'].iloc[0]
            pred_df['away_avg_corners_for'] = away_row['Avg Corners For'].iloc[0]
            pred_df['away_avg_corners_against'] = away_row['Avg Corners Against'].iloc[0]

        pred_df_model = pred_df.drop(columns=["HomeTeam", "AwayTeam"])

        lambda_home = model_home.predict(pred_df_model)[0]
        lambda_away = model_away.predict(pred_df_model)[0]
        poisson_pred = lambda_home + lambda_away
        home_win, draw_prob, away_win, btts_prob, over_2_5_prob = predict_match_outcomes(lambda_home, lambda_away)
        prob_handicap = predict_asian_handicap(lambda_home, lambda_away, handicap)

        if 'HC' in df.columns and 'AC' in df.columns:
            model_corners_home, model_corners_away = fit_poisson_models_corners(df_form)
            lambda_corner_home = model_corners_home.predict(pred_df_model)[0]
            lambda_corner_away = model_corners_away.predict(pred_df_model)[0]
            total_corners = lambda_corner_home + lambda_corner_away
            prob_over, prob_under = calculate_corners_probabilities(total_corners, threshold=corners_threshold)
            corners_message = f"Escanteios: {total_corners:.2f} (Prob. Over {corners_threshold}: {prob_over:.2%} | Odds: {odds(prob_over):.2f})"
        else:
            corners_message = "Dados de escanteios não disponíveis"

        gb_model, gb_features = train_gradient_boosting_model(df_form)
        gb_pred = gb_model.predict(pred_df_model[gb_features])[0]

        ensemble_pred = ensemble_prediction(poisson_pred, gb_pred, weight=0.5)

        max_goals_val = 6
        probs = np.zeros((max_goals_val+1, max_goals_val+1))
        for i in range(max_goals_val+1):
            for j in range(max_goals_val+1):
                probs[i, j] = poisson.pmf(i, lambda_home) * poisson.pmf(j, lambda_away)
        heatmap = go.Figure(data=go.Heatmap(z=probs, x=list(range(max_goals_val+1)), y=list(range(max_goals_val+1)),
                                            colorscale='Viridis', text=np.round(probs, 3), hoverinfo='text'))
        heatmap.update_layout(title=f"Heatmap de Gols: {home_team} vs {away_team}",
                              xaxis_title="Gols Visitante", yaxis_title="Gols Casa")

        message = f"""
{warning}**Análise de Jogo: {home_team} vs {away_team}**

- **Previsão de Gols (Poisson):**
  - {home_team}: {lambda_home:.2f} gols
  - {away_team}: {lambda_away:.2f} gols
  - Total Previsto (Poisson): {poisson_pred:.2f} gols

- **Previsão de Gols (Gradient Boosting):** {gb_pred:.2f} gols

- **Previsão de Gols (Ensemble):** {ensemble_pred:.2f} gols

- **Previsão de Escanteios:** {corners_message}

- **Probabilidades de Resultado:**
  - Vitória {home_team}: {home_win:.2%}
  - Empate: {draw_prob:.2%}
  - Vitória {away_team}: {away_win:.2%}

- **Outros Eventos:**
  - BTTS: {btts_prob:.2%} (Odds: {odds(btts_prob):.2f})
  - Over 2.5 Gols: {over_2_5_prob:.2%} (Odds: {odds(over_2_5_prob):.2f})
  - Prob. de Handicap {handicap}: {prob_handicap:.2%} (Odds: {odds(prob_handicap):.2f})

- **Análise de Forma (Últimos {best_window} jogos):**
  - {home_team}: Média de gols marcados: {compute_rolling_features(df, home_team, best_window)['home']['rolling_GF']:.2f}, sofridos: {compute_rolling_features(df, home_team, best_window)['home']['rolling_GA']:.2f}
  - {away_team}: Média de gols marcados: {compute_rolling_features(df, away_team, best_window)['away']['rolling_GF']:.2f}, sofridos: {compute_rolling_features(df, away_team, best_window)['away']['rolling_GA']:.2f}

- **Desempenho contra Adversários Similares a {away_team}:**
  - {similar_opponent_performance(df, home_team, classification_table, reference_team=away_team)}
        """

        odds_data_calculated = [
            {"Evento": "Vitória Mandante", "Probabilidade": f"{home_win:.2%}", "Odds Calculadas": f"{odds(home_win):.2f}",
             "Recomendação": "Aposta recomendada" if home_win > prob_threshold else "Cautela"},
            {"Evento": "Empate", "Probabilidade": f"{draw_prob:.2%}", "Odds Calculadas": f"{odds(draw_prob):.2f}",
             "Recomendação": "Atenção"},
            {"Evento": "Vitória Visitante", "Probabilidade": f"{away_win:.2%}", "Odds Calculadas": f"{odds(away_win):.2f}",
             "Recomendação": "Cautela"},
            {"Evento": "BTTS", "Probabilidade": f"{btts_prob:.2%}", "Odds Calculadas": f"{odds(btts_prob):.2f}",
             "Recomendação": "Aposta recomendada" if btts_prob > prob_threshold else "Cautela"},
            {"Evento": "Over 2.5 Gols", "Probabilidade": f"{over_2_5_prob:.2%}", "Odds Calculadas": f"{odds(over_2_5_prob):.2f}",
             "Recomendação": "Aposta recomendada" if over_2_5_prob > prob_threshold else "Cautela"},
            {"Evento": f"Handicap {handicap}", "Probabilidade": f"{prob_handicap:.2%}", "Odds Calculadas": f"{odds(prob_handicap):.2f}",
             "Recomendação": "Aposta recomendada" if prob_handicap > prob_threshold else "Cautela"}
        ]

        performance_table = performance_by_opponent_level(df, home_team, classification_table, margin=3)
        similarity_table = compute_similarity_table(df, home_team, classification_table, reference_team=away_team)

        # Enviar notificação ao Telegram (opcional)
        if home_win > prob_threshold or btts_prob > prob_threshold or over_2_5_prob > prob_threshold:
            telegram_msg = f"Recomendação de aposta: {home_team} vs {away_team}\n" \
                           f"Vitória Casa: {home_win:.2%}\nBTTS: {btts_prob:.2%}\nOver 2.5: {over_2_5_prob:.2%}"
            send_telegram_notification(telegram_msg, TELEGRAM_CHAT_ID, TELEGRAM_BOT_TOKEN)

        return dcc.Markdown(message, dangerously_allow_html=True), odds_data_calculated, similarity_table.to_dict('records'), heatmap, performance_table.to_dict('records')
    except Exception as ex:
        logging.error(f"Erro no callback analyze_match: {ex}")
        return dcc.Markdown(f"Erro ao gerar análise: {ex}", dangerously_allow_html=True), no_update, no_update, go.Figure(), no_update

@app.callback(
    Output('home-goals-trend-graph', 'figure'),
    [Input('league-dropdown', 'value'),
     Input('home-team-dropdown', 'value')]
)
def update_home_goals_trend_graph(selected_league: str, home_team: str):
    if not home_team:
        return go.Figure()
    df = load_data(selected_league)
    home_df = df[df['HomeTeam'] == home_team].sort_values('Date')
    if home_df.empty:
        return go.Figure()

    fig = go.Figure()
    fig.add_trace(go.Scatter(x=home_df['Date'], y=home_df['FTHG'], mode='lines+markers', name='Gols Marcados'))
    fig.add_trace(go.Scatter(x=home_df['Date'], y=home_df['FTAG'], mode='lines+markers', name='Gols Sofridos'))
    fig.update_layout(
        title=f"Tendência de Gols - {home_team} (Mandante)",
        xaxis_title="Data",
        yaxis_title="Gols",
        legend_title="Legenda"
    )
    return fig

@app.callback(
    Output('away-goals-trend-graph', 'figure'),
    [Input('league-dropdown', 'value'),
     Input('away-team-dropdown', 'value')]
)
def update_away_goals_trend_graph(selected_league: str, away_team: str):
    if not away_team:
        return go.Figure()
    df = load_data(selected_league)
    away_df = df[df['AwayTeam'] == away_team].sort_values('Date')
    if away_df.empty:
        return go.Figure()

    fig = go.Figure()
    fig.add_trace(go.Scatter(x=away_df['Date'], y=away_df['FTAG'], mode='lines+markers', name='Gols Marcados'))
    fig.add_trace(go.Scatter(x=away_df['Date'], y=away_df['FTHG'], mode='lines+markers', name='Gols Sofridos'))
    fig.update_layout(
        title=f"Tendência de Gols - {away_team} (Visitante)",
        xaxis_title="Data",
        yaxis_title="Gols",
        legend_title="Legenda"
    )
    return fig

@app.callback(
    Output('form-comparison-graph', 'figure'),
    [Input('league-dropdown', 'value'),
     Input('home-team-dropdown', 'value'),
     Input('away-team-dropdown', 'value')]
)
def update_form_comparison_graph(selected_league: str, home_team: str, away_team: str):
    if not home_team or not away_team:
        return go.Figure()
    df = load_data(selected_league)
    window = optimize_window_size(df)
    home_form = compute_rolling_features(df, home_team, window=window)['home']
    away_form = compute_rolling_features(df, away_team, window=window)['away']

    fig = go.Figure(data=[
        go.Bar(name='Gols Marcados', x=[home_team, away_team], y=[home_form['rolling_GF'], away_form['rolling_GF']]),
        go.Bar(name='Gols Sofridos', x=[home_team, away_team], y=[home_form['rolling_GA'], away_form['rolling_GA']])
    ])
    fig.update_layout(
        barmode='group',
        title=f"Comparação de Forma Recente (Últimos {window} Jogos)",
        xaxis_title="Times",
        yaxis_title="Média de Gols",
        legend_title="Legenda"
    )
    return fig

@app.callback(
    Output("download-odds", "data"),
    [Input("export-odds-btn", "n_clicks")],
    [State('odds-table', 'data'),
     State('home-team-dropdown', 'value'),
     State('away-team-dropdown', 'value')]
)
def export_odds(n_clicks, odds_table_data, home_team, away_team):
    if n_clicks:
        odds_df = pd.DataFrame(odds_table_data)
        date_str = datetime.now().strftime("%Y%m%d")
        filename = f"odds_{home_team}_vs_{away_team}_{date_str}.csv"
        return dict(content=odds_df.to_csv(index=False), filename=filename)
    return no_update

@app.callback(
    Output("download-similarity", "data"),
    [Input("export-similarity-btn", "n_clicks")],
    [State('similarity-table', 'data'),
     State('home-team-dropdown', 'value'),
     State('away-team-dropdown', 'value')]
)
def export_similarity(n_clicks, similarity_table_data, home_team, away_team):
    if n_clicks:
        similarity_df = pd.DataFrame(similarity_table_data)
        date_str = datetime.now().strftime("%Y%m%d")
        filename = f"similarity_{home_team}_vs_{away_team}_{date_str}.csv"
        return dict(content=similarity_df.to_csv(index=False), filename=filename)
    return no_update

# =============================================================================
# INTEGRAÇÃO COM NGROK E EXECUÇÃO DO APP
# =============================================================================
if __name__ == '__main__':
    # Libera a porta 8051 (ajuste o comando se estiver em outro sistema operacional)
    print("Liberando a porta 8051...")
    os.system("fuser -k 8051/tcp")

    # Configure o authtoken do ngrok (substitua pelo seu token válido)
    ngrok.set_auth_token("2uckAKvhBja7ftNu54eA2VJLi73_7Bu1Qepb5jq3MkfwV7euJ")

    # Encerra quaisquer túneis ngrok existentes para evitar conflitos
    ngrok.kill()

    # Função para rodar o servidor Dash em uma thread separada
    def run_dash():
        print("Iniciando servidor Dash na porta 8051...")
        app.run(host='0.0.0.0', port=8051, debug=False)

    # Inicia o servidor Dash em uma thread
    thread = Thread(target=run_dash)
    thread.start()

    # Aguarda o servidor iniciar
    time.sleep(5)

    # Cria o túnel ngrok para a porta 8051
    try:
        public_url = ngrok.connect(8051)
        print(f"Dashboard disponível em: {public_url}")
    except Exception as e:
        print(f"Erro ao criar o túnel ngrok: {e}")
```
